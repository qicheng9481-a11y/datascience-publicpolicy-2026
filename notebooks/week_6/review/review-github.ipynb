# Week 6 Review: Demystifying Your Programming Tools
### Understanding Git, GitHub, VS Code, Anaconda, GitHub Desktop, and GitHub Web

When starting out with data science for public policy and economics, one of the biggest hurdles is not actually the Python code itself. Instead, it is understanding the massive collection of software, terms, and jargon that comes with it.

If you have ever felt confused about the difference between Git and GitHub, or wondered why you need both VS Code and Anaconda, this notebook is for you! Let us break down exactly what each tool does and how they all talk to one another.

## 1. The Metaphor: The Policy Research Lab

To make sense of all these pieces, let us imagine you are setting up a professional **Policy Research Lab** to analyze economic datasets and write public policy recommendations. Here is how each software relates to your lab:

| Software Component | Lab Analogy | What It Actually Does |
| :--- | :--- | :--- |
| **Python** | **Raw Economic Principles & Data** | The fundamental mathematical rules and logic you use to analyze data and build models (e.g., calculating inflation or regression). |
| **Anaconda** | **The Research Software Suite** | A pre-stocked suite of tools. Instead of downloading Python, Jupyter Notebooks, and statistical toolkits individually, Anaconda installs them all in one package so you are ready to work immediately. |
| **VS Code (Visual Studio Code)** | **The Analyst's Desk & Notebook** | A clean, comfortable desk where you lay out your documents, organize files, write your reports, and run calculations. It is a nice workspace, but it does not create the data or tools on its own. |
| **Git** | **The Legislative Audit Log** | An automatic audit log. Every time you make a significant change to a policy draft or clean a dataset, Git takes a permanent snapshot. If you make a mistake or corrupt a file, you can roll back your project to a previous snapshot perfectly. |
| **GitHub Web** | **The Global Policy Archive (Cloud)** | An online public archive on the internet where you upload backups of your policy reports and datasets. This allows you to show off your work to the public, share it with advisors, or collaborate with researchers around the world. |
| **GitHub Desktop** | **The Archive Control Panel** | A simple remote control with clear buttons (like "Commit" and "Push"). It lets you take your snapshots and sync them to your online archive without needing to type complex database commands. |

## 2. Core Vocabulary: Cutting Through the Jargon

When people talk about version control and GitHub, they use specialized terms. Here is a handy translation guide:

| Jargon Term | What It Actually Means in Plain English |
| :--- | :--- |
| **Repository (or "Repo")** | A fancy word for your project's main folder. It is just a folder on your computer that Git is keeping an eye on. |
| **Local** | Refers to files and actions stored directly on your physical computer's hard drive. |
| **Remote** | Refers to files and actions stored on the GitHub servers on the internet (in the cloud). |
| **Commit** | Taking a permanent snapshot of your project's current state. It saves a history of exactly what changed. |
| **Push** | Sending your local snapshots (commits) up to the GitHub cloud so the remote repository matches your computer. |
| **Pull** | Downloading any new changes from the GitHub cloud down to your local computer. |
| **Clone** | Downloading a complete copy of a repository from GitHub to your computer for the very first time. |
| **Fork** | Creating a personal copy of someone else's GitHub project inside your own GitHub account so you can experiment with it. |

## 3. How to Run Code and Show GitHub Changes

A common question is: **How do I write and run code, and then see the changes on GitHub?** here is the step by step process:

### Step 1: Open VS Code and Run Code
1. Open your project folder in VS Code.
2. Open a Jupyter Notebook file (like this one).
3. To run a code cell, click the **Play** button next to the cell, or press **Shift + Enter** on your keyboard.
4. Make some changes (e.g., write a new variable or print a new string).
5. Save your notebook file by pressing **Cmd + S** (Mac) or **Ctrl + S** (Windows).

### Step 2: Open GitHub Desktop to See the Changes
1. Open the **GitHub Desktop** application.
2. On the left sidebar under the **Changes** tab, you will immediately see the notebook file you just saved.
3. Click on the notebook file. GitHub Desktop will display a "diff" (difference), which highlights the exact modifications you made (red for deleted content, green for new code or output).

### Step 3: Save a Snapshot (Commit)
1. At the bottom left of GitHub Desktop, you will see a box to enter a commit message.
2. Write a short, descriptive message (e.g., "Update G7 country list example" or "Fix calculation formula").
3. Click the blue **Commit to main** button. This saves a permanent snapshot on your physical laptop (local).

### Step 4: Sync to the Internet (Push)
1. Click the **Push origin** button at the top of the screen.
2. This uploads your new snapshot to the online repository on GitHub Web (remote). Now your work is backed up and visible to your instructors and teammates!
